## 2. Data Partitioning

### Split into training, validation and test dataset

- **Training Dataset**: The sample of data used to fit the model.
    
- **Validation Dataset**: The sample of data used to provide an unbiased evaluation of a model fit on the training dataset while tuning model hyperparameters. The evaluation becomes more biased as skill on the validation dataset is incorporated into the model configuration.

- **Test Dataset**: The sample of data used to provide an unbiased evaluation of a final model fit on the training dataset.

```python
# Create X values
X = df.drop('TargetB', axis=1)
print(X)
```

```python
# Create Y values
y = df['TargetB']
print(y)
```

[Encoding Categorical Features](https://towardsdatascience.com/encoding-categorical-features-21a2651a065c)

```python
# Instantiate LabelEncoder object
le = LabelEncoder()

# Apply le on categorical feature columns
X[['StatusCat96NK', 'DemGender']] = X[['StatusCat96NK', 'DemGender']].apply(lambda col: le.fit_transform(col))
X
```

```python
# Split into training set and testing set
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.2, random_state = 1)
```

```python
# Split into training set and validation set
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size = 0.25, random_state=1)
```

```python
# How many rows in each set
print("Training set data have {} of records, which is {}%.".format(
    X_train.shape[0], (len(X_train)/len(df)) * 100))
print("Validation set data have {} of records, which is {}%.".format(
    X_val.shape[0], (len(X_val)/len(df)) * 100))
print("Testing set data have {} of records, which is {}%.".format(
    X_test.shape[0], (len(X_test)/len(df)) * 100))
```

### K-fold Cross Validation

Cross-validation is a resampling procedure used to evaluate machine learning models on a limited data sample.

**K-Fold CV** is where a given data set is split into a K number of sections/folds where each fold is used as a testing set at some point. Lets take the scenario of 5-Fold cross validation(K=5). Here, the data set is split into 5 folds. In the first iteration, the first fold is used to test the model and the rest are used to train the model. In the second iteration, 2nd fold is used as the testing set while the rest serve as the training set. This process is repeated until each fold of the 5 folds have been used as the testing set.


```python
decision_tree_cv = cross_val_score(DecisionTreeClassifier(), X, y, cv = 5)
print("Mean cross-validation score: {}".format(np.mean(decision_tree_cv)))
```

**Pros of K-Fold CV:**
- Due to sample variability between training and test set, our model gives a better prediction on training data but fail to generalize on test data. This leads to a low training error rate but a high test error rate. This problem can be solved using cross validation to **reduce bias**

**Cons of K-Fold CV:**
- The training algorithm is **computationally intensive** as the algorithm has to be rerun from scratch k times.

## 3. Decision Tree

### 3.1 Build a Decision Tree classifier

```python
# Decision tree classifier
decision_tree = DecisionTreeClassifier(random_state = 0).fit(np.asmatrix(X_train), y_train)
```

### 3.2 Performance of the model against training data

```python
decision_tree.score(X_train, y_train)
```

### 3.3 Performance of the model against validation data

```python
decision_tree.score(X_val, y_val)
```

### 3.4 Performance of the model against test data

```python
decision_tree.score(X_test, y_test)
```

*Predict using training data tends overfitting whereas test data don't.*

### 3.5 Performance Metrics of model

```python
y_pred = decision_tree.predict(X_test)
print(classification_report(y_test, y_pred))
```

### 3.6 Important Features
